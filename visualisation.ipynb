```python
data_clean['CLIENT_AGE'] = data_clean['DAYS_BIRTH']/-365
data_clean['CLIENT_AGE']
```

```python
age = data_clean['CLIENT_AGE']
```

```python
fig, ax = plt.subplots(figsize = (15,8))


age.plot(kind = "hist", density = True, alpha = 0.65, bins = 100) 
   
age.plot(kind = "kde")



ax.set_xlabel("Age")
   
x_start, x_end = 0, 100
ax.set_xlim(x_start, x_end)


ax.set_ylim(0, 0.04)
ax.set_yticklabels([])
ax.set_ylabel("Frequency")

median=age.median()
ax.axvline(median, color='green', linestyle='-', label="Median")
mean=age.mean()
ax.axvline(mean, color='black', linestyle=':', label="Mean")

ax.legend()


ax.grid(False)
ax.set_title("Age Histogram", size = 17, pad = 10)

plt.show()
```
