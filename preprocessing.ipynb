Feature Engineering

Time features are converted from days to years

```python
df['CLIENT_AGE'] = df['DAYS_BIRTH']/-365
df['CLIENT_AGE']
```

```python
df['EMPLOYMENT_YEARS'] = df['DAYS_EMPLOYED']/-365
df['EMPLOYMENT_YEARS']
```

```python
df['REG_YEARS'] = df['DAYS_REGISTRATION']/-365
df['REG_YEARS']
```

```python
df['ID_AGE'] = df['DAYS_ID_PUBLISH']/-365
df['ID_AGE']
```

```python
df['PHONE_AGE'] = df['DAYS_LAST_PHONE_CHANGE']/-365
df['PHONE_AGE']
```

```python
df = df.drop(['DAYS_BIRTH', 'DAYS_EMPLOYED', 'DAYS_REGISTRATION', 'DAYS_ID_PUBLISH', 'DAYS_LAST_PHONE_CHANGE'], axis = 1)
```

```python
df
```

HOUR_APPR_PROCESS_START is turned from 24 hour time to four times of day

```python
df['HOUR_APPR_PROCESS_START'].unique()
```

```python
map_dict ={18: 'EVENING', 19: 'EVENING', 20: 'EVENING', 21: 'EVENING',
           22: 'EVENING', 23: 'NIGHT', 0: 'NIGHT', 1: 'NIGHT', 
           2: 'NIGHT', 3: 'NIGHT', 4: 'NIGHT', 5: 'NIGHT', 6: 'MORNING',
           7: 'MORNING', 8: 'MORNING', 9: 'MORNING', 10: 'MORNING',
           11: 'MORNING', 12: 'AFTERNOON', 13: 'AFTERNOON', 14: 'AFTERNOON',
           15: 'AFTERNOON', 16: 'AFTERNOON', 17: 'AFTERNOON'}
df['TIME_OF_DAY'] = df.HOUR_APPR_PROCESS_START.map(map_dict)
```

```python
df['TIME_OF_DAY'].unique()
```

```python
df = df.drop(['HOUR_APPR_PROCESS_START'], axis=1)
df 
```
