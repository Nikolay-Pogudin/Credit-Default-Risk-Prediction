AMT_ANNUITY

```python
application['Interest_Rate'] = application['AMT_ANNUITY']/application['AMT_CREDIT'] * 100
application['Interest_Rate']
#keep this to use as a feature in analysis
```

```python
application['Interest_Rate'].isnull().sum(axis = 0)
```

```python
application['Interest_Rate'].mean()
```

```python
application['Interest_Rate'] = application['Interest_Rate'].fillna((application['Interest_Rate'].mean()))
```

```python
application['Interest_Rate'].isnull().sum(axis = 0)
```

```python
application['Interest_Rate'].mean()
```

```python
application["AMT_ANNUITY"].isnull().sum(axis = 0)
```

```python
application.info(verbose=True, null_counts=True)
```

```python
application["AMT_ANNUITY_FULL"] = application['AMT_CREDIT'] * application['Interest_Rate']
application["AMT_ANNUITY_FULL"] = application["AMT_ANNUITY_FULL"]/100
application["AMT_ANNUITY_FULL"]
```

```python
application['AMT_ANNUITY_FULL'].isnull().sum(axis = 0)
```

```python
application['AMT_ANNUITY_FULL'] = application['AMT_ANNUITY_FULL'].round(decimals = 2)
application['AMT_ANNUITY_FULL']
```

Below a check is performed to ensure AMT_ANNUITY_FULL is the same as AMT_ANNUITY

```python
pd.set_option('display.max_rows', 130)
data_compare = [application['AMT_ANNUITY_FULL'], application['AMT_ANNUITY']]
df_compare = pd.concat(data_compare, axis=1)
df_compare['new'] = (df_compare['AMT_ANNUITY_FULL'] == df_compare['AMT_ANNUITY']) 


df_compare.head()
df_compare[df_compare['new'] == False]
```

```python
df_compare.head()
df_compare[df_compare['new'] == True]
```
